```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

def find_similar_stores(df, trial_stores, top_pct=5):
    """
    Find the most similar stores to the trial stores based on monthly sales data.
    Uses both Pearson correlation and magnitude distance to calculate similarity.
    
    Parameters:
        df (pd.DataFrame): The input data with columns: MONTH, STORE_NBR, TOTAL_SALES, TOTAL_QUANTITY, NUM_TRANSACTIONS.
        trial_stores (list): List of trial store numbers to compare.
        top_pct (int): Percentage of top similar stores to return (default: 10%).
    
    Returns:
        dict: A dictionary where keys are trial store numbers and values are DataFrames of top similar stores.
    """
    results = {}
    
    for trial in trial_stores:
        # Get data for the trial store
        trial_data = df[df['STORE_NBR'] == trial]
        trial_months = trial_data['MONTH'].unique()
        
        if len(trial_months) == 0:
            print(f"No data found for trial store {trial}")
            continue
        
        # Get all other stores (excluding trial stores)
        other_stores = df[~df['STORE_NBR'].isin(trial_stores)]['STORE_NBR'].unique()
        
        similarity_scores = []
        
        for store in other_stores:
            # Get data for the other store
            store_data = df[
                (df['STORE_NBR'] == store) & 
                (df['MONTH'].isin(trial_months))
            ].sort_values('MONTH')
            
            # Skip if the store doesn't have data for the same months
            if len(store_data) != len(trial_data):
                continue
            
            # Align months between trial and other store
            trial_sorted = trial_data.sort_values('MONTH')
            store_sorted = store_data.sort_values('MONTH')
            
            if not (trial_sorted['MONTH'].values == store_sorted['MONTH'].values).all():
                continue
            
            # Calculate Pearson correlation for each metric and average them
            corrs = []
            for col in ['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']:
                corr, _ = pearsonr(trial_sorted[col], store_sorted[col])
                corrs.append(corr)
            pearson_score = np.mean(corrs)
            
            # Normalize data for distance calculation
            trial_normalized = (trial_sorted[['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']] - 
                               trial_sorted[['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']].mean()) / \
                               trial_sorted[['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']].std()
            store_normalized = (store_sorted[['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']] - 
                                store_sorted[['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']].mean()) / \
                                store_sorted[['TOTAL_SALES', 'TOTAL_QUANTITY', 'NUM_TRANSACTIONS']].std()
            
            # Calculate Euclidean distance
            distance = np.linalg.norm(trial_normalized.values - store_normalized.values)
            
            # Convert distance to similarity score (smaller distance = more similar)
            distance_score = 1 / (1 + distance)  # Inverse of distance (range: 0 to 1)
            
            # Combine Pearson and distance scores (weighted average)
            combined_score = (pearson_score + distance_score) / 2
            
            similarity_scores.append((store, combined_score))
        
        # Sort stores by combined similarity score (descending)
        similarity_scores.sort(key=lambda x: x[1], reverse=True)
        
        # Select top percentage of similar stores
        num_top = 5
        top_stores = similarity_scores[:num_top] 
        # Store results in a DataFrame
        results_df = pd.DataFrame(top_stores, columns=['STORE_NBR', 'SIMILARITY_SCORE'])
        results[trial] = results_df
    
    return results
```

```python
# Find similar stores
trial_stores = [88,77,86]
similar_stores = find_similar_stores(monthly_store_insights, trial_stores, top_pct=10)

# Print results
for trial, result in similar_stores.items():
    print(f"Trial Store {trial}:")
    print(result)
```

```python
filtered_df = df[
    (df['STORE_NBR'].isin([167, 35, 41,77,88,229,159,188,86,147,231]))  &  # Filter for Store No. 167
    (df['year'] == 2019) &  # Filter for year 2019
    (df['MONTH']>= 2) &    # Filter for months >= February
    (df['MONTH']<= 4)      # Filter for months <= April\
]
filtered_df
```

```python
grouped = filtered_df.groupby('STORE_NBR').agg(
    Total_Sales=('TOT_SALES', 'sum'),  # Total sales per store
    Number_of_Purchasing_Customers=('LYLTY_CARD_NBR', 'nunique'),  # Unique customers per store
    Total_Purchases=('LYLTY_CARD_NBR', 'count')  # Total purchases per store
)

# Calculate metrics
grouped['Average_Sales'] = grouped['Total_Sales'] / grouped['Number_of_Purchasing_Customers']
grouped['Average_Purchases_per_Customer'] = grouped['Total_Purchases'] / grouped['Number_of_Purchasing_Customers']
grouped
# Rename columns for clarity
grouped = grouped.rename(columns={
    'Total_Sales': 'Total Sales',
    'Number_of_Purchasing_Customers': 'Number of Purchasing Customers',
    'Total_Purchases': 'Total Purchases',
    'Average_Sales': 'Average Sales per Customer',
    'Average_Purchases_per_Customer': 'Average Purchases per Customer'
})
grouped = grouped.reset_index()
grouped
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Define Trial and Control Store pairs
trial_control_pairs = {
    "77": [35, 41, 77, 167],  # Store 77 with its control stores
    "88": [88, 229],          # Store 88 with its control store
    "86": [86, 147]           # Store 86 with its control store
}

# Metrics to visualize
metrics = ["Total Sales", "Number of Purchasing Customers", "Average Sales per Customer"]

# Set up multiple subplots in a single grid layout
fig, axes = plt.subplots(len(trial_control_pairs), len(metrics), figsize=(15, 10))  # Adjust size to fit all

# Flatten axes for easy indexing if multiple rows exist
axes = axes.flatten()

# Track subplot index
index = 0

# Generate visualizations for each Trial Store with its specific Control Stores
for trial, stores in trial_control_pairs.items():
    subset = grouped[grouped["STORE_NBR"].isin(stores)]  # Filter relevant stores

    for metric in metrics:
        sns.barplot(data=subset, x="STORE_NBR", y=metric, palette="viridis", ax=axes[index])
        axes[index].set_title(f"Store {trial} vs Control Stores: {metric}", fontsize=12, fontweight="bold")
        axes[index].set_xlabel("Store Number")
        axes[index].set_ylabel(metric)
        axes[index].tick_params(axis='x', rotation=45)  # Rotate x-labels for readability
        index += 1  # Move to next subplot

# Adjust layout for better spacing
plt.tight_layout()

# Display the visualization only once
plt.show()
```
